```python
import matplotlib as plt
from sympy import diff, Eq, solve, plot
from sympy.abc import x
init_printing()
```

```python
# our function
d0 = -x**2 + 3*x - 2

# Y @ endpoints of [1, 3]
d0.subs({'x': 1}), d0.subs({'x': 3})
```

```python
# first derivative
d1 = diff(d0)
d1
```

```python
# solve first derivative to find critical points
solve(Eq(d1, 0))
```

```python
plot(d0, xlim=(0, 3.5), ylim=(-5, 1))
```

```python
# value of Y @ critical point of 3/2
d0.subs({'x': 3/2})
```

```python
print('''
    x    |    f(x)    |  description
==================================================
    1    |    0       |
    3/2  |    1/4     | absolute maxima
    3    |    -2      | absolute minima (endpoint)
''')
```
